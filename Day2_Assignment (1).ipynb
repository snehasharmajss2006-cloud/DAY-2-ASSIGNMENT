```python
import pandas as pd
```

```python
# Creating a dataset of popular Indian singers and their achievements
data = { 'Singer': ['Arijit Singh', 'Shreya Ghoshal', 'Sonu Nigam', 'Lata Mangeshkar', 'Kishore Kumar'], 
         'Years_Active': [15, 20, 30, 70, 40], 
         'Number_of_Songs': [500, 700, 600, 1000, 800],
         'Awards_Won': [50, 60, 40, 100, 80] }
```

```python
# Convert the dictionary into a Pandas DataFrame 
singers_df = pd.DataFrame(data)
```

```python
# Display basic information about the dataset
singers_df.info()
```

```python
# Calculate basic statistical metrics 
singers_df.describe()

```

```python
# Calculate specific statistics
max_songs = singers_df['Number_of_Songs'].max()
min_songs = singers_df['Number_of_Songs'].min() 
mean_awards = singers_df['Awards_Won'].mean()
median_years = singers_df['Years_Active'].median()
```

```python
print(f"Maximum number of songs: {max_songs}" )
print(f"Minimum number of songs: {min_songs}")
print(f"Average awards won: {mean_awards:.2f}") 
print(f"Median years active: {median_years}")
```

```python
# Sort singers by the number of songs 
sorted_singers = singers_df.sort_values(by='Awards_Won', ascending=False)
sorted_singers
```
